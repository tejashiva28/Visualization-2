```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Load Excel file into a DataFrame
df = pd.read_excel("C:/Users/tejas/Downloads/IncidentVsNumber.xlsx")

# Check for null values in the DataFrame
null_values = df.isnull().sum()
print("Null Values:")
print(null_values)

# Drop rows with null values in specific columns (e.g., 'Incident Types' and 'Number of Incidents')
columns_to_check = ['Incident Types', 'Number of Incidents']
df = df.dropna(subset=columns_to_check)

df = df[df['Incident Types'] != 'Grand Total']

# Pie Chart
plt.figure(figsize=(8, 8))
plt.pie(df['Number of Incidents'], labels=df['Incident Types'], autopct='%1.1f%%')
plt.title('Distribution of Incident Types')
plt.show()

# Bar Graph
plt.figure(figsize=(10, 6))
plt.bar(df['Incident Types'], df['Number of Incidents'])
plt.xlabel('Incident Types')
plt.ylabel('Number of Incidents')
plt.title('Incidents by Type')
plt.xticks(rotation=45, ha='right')
plt.show()

# Line Chart
plt.figure(figsize=(10, 6))
plt.plot(df['Incident Types'], df['Number of Incidents'], marker='o')
plt.xlabel('Incident Types')
plt.ylabel('Number of Incidents')
plt.title('Trend of Incidents')
plt.xticks(rotation=45, ha='right')
plt.grid(True)
plt.show()

# Box Plot
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['Incident Types'], y=df['Number of Incidents'])
plt.xlabel('Incident Types')
plt.ylabel('Number of Incidents')
plt.title('Distribution of Incidents')
plt.xticks(rotation=45, ha='right')
plt.show()

# Interactive Bar Graph (Plotly)
fig = px.bar(df, x='Incident Types', y='Number of Incidents', title='Incidents by Type')
fig.update_layout(xaxis={'categoryorder':'total descending'})
fig.show()
```
